```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

def get52WeekRange(ticker):
    quote_table = get_quote_table(ticker)
    s = quote_table['52 Week Range']
    l = s.split('-')
    return float(l[0].strip()),float(l[1].strip())

def getCurrentPrice(ticker):
    return get_live_price(ticker)

def getCompanyName(ticker):    
    name = ""
    URL = 'http://finance.yahoo.com/q/ks?s='+ticker
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    container = soup.find('h1', {"data-reactid":'7'})
    if container:      
        name = container.text
    return name

def getStockInfo(stockArr):
    for individualStock in stockArr:       
        pegratio = getPEGRatio(individualStock)
        try:
            if (type(pegratio) == str):
              pegvalue = float(pegratio)
            else:
              pegvalue =   pegratio

            currprice = getCurrentPrice(individualStock)
            company = getCompanyName(individualStock)
            low, high = get52WeekRange(individualStock)

            col_companyName.append(company)
            col_ticker.append(individualStock)
            col_pegRatio.append(pegvalue)
            col_currentPrice.append(currprice)
            col_52High.append(high)
            col_52Low.append(low)
            col_DiffFromLowest.append(currprice - low)
            col_DiffFromHighest.append(high - currprice)

            if pegvalue <= 2.0:                  
                toprint = "stock:\t\t" + individualStock + " PEG**:"+ str(pegvalue) + " CurrentPrice:"+ str(currprice)
                print("\x1b[31m\"" + toprint +  "\"\x1b[0m")                    
            else:
                print("stock:\t\t" + individualStock + " PEG:"+ str(pegvalue) + " CurrentPrice:"+ str(currprice))          
        except ValueError:        
            print("ValueError")

def getPEGRatio(ticker):
    dfstats = get_stats_valuation(ticker)
    return dfstats.iloc[4][1] #peg ratio based on current date

#https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html
def color_matchstr(matchstr):
    if 'watch' in matchstr:
        color = 'yellow'   
    else:
        color = 'white'
    return 'background-color: %s' % color    

def getCurrentPERatio(ticker):
  stats = get_quote_table(ticker)
  print(stats)
  return stats['PE Ratio (TTM)']


#get minimum, maximum and avg PE ratio
#https://ycharts.com/companies/CSCO/pe_ratio
def getHistoricalPERatio(ticker):    
    name = ""
    URL = 'https://ycharts.com/companies/'+ticker +'/pe_ratio'
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    #mytable  = soup.find("table", {"class": "rangeModTable"})
    mytable = soup.findAll('table', attrs={'class': 'rangeModTable'})    
    data = {}
    my_table = soup.find("table", attrs={"class": "rangeModTable"})
    my_table_data = my_table.find_all("tr")
    #for trd in my_table_data:
    #  print(trd)
    headings = []
    dictValues = {}
    for i in range(len(my_table_data)):
      for td in my_table_data[i].find_all("td"):
        #print(td)
        # remove any newlines and extra spaces from left and right
        headings.append(td.text.replace('\n', ' ').strip())
        #print(headings)
        #['Minimum', '11.89', 'Mar 18 2020', 'Maximum', '41.85', 'Apr 15 2019', 'Average', '25.12', '']
        #1, 4 , 7
        if (len(headings) == 8):
          dictValues['Minimum'] = headings[1]
          dictValues['Maximum'] = headings[4]
          dictValues['Average'] = headings[7]
    return dictValues


# functions to get and parse data from FinViz

def fundamental_metric(soup, metric):
    return soup.find(text = metric).find_next(class_='snapshot-td2').text

def get_fundamental_data(df):
  for symbol in df.index:
    try:
        url = ("https://finviz.com/quote.ashx?t=" + symbol.lower())
        soup = BeautifulSoup(requests.get(url).content) 
        for m in df.columns:                
            df.loc[symbol,m] = fundamental_metric(soup,m)                
    except Exception as e:
        print (e)
  return df

    
```
